```python
'''
Author: Aditya Vivek Khatode
title: Deep Learning
'''
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import keras
from keras.layers import Dense, Activation,Dropout
from keras.models import Sequential
```

```python
name= ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'PRICE']
df = pd.read_csv(filepath_or_buffer="housing.csv",delim_whitespace=True,names=name)
df.head()
```

```python
print(df.shape)
```

```python
df.describe()
```

```python
# Splitting the dependent feature and independent feature
X = df.iloc[:,:-1]
y= df.PRICE
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 4)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Scaling the dataset
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
model = Sequential()

model.add(Dense(128,activation  = 'relu',input_dim =13))
model.add(Dense(64,activation  = 'relu'))
model.add(Dense(32,activation  = 'relu'))
model.add(Dense(16,activation  = 'relu'))
model.add(Dense(1))
model.compile(optimizer = 'adam',loss = 'mean_squared_error')
```

```python
model.fit(X_train, y_train, epochs = 100)
```

```python
new_input=np.array([[0.00632,18.0,2.31,0,0.538,6.575,65.2,4.0900,1,296.0,15.3,396.90,4.98]])
new_input=sc.transform(new_input)
prediction=model.predict(new_input)
print(prediction)
```
